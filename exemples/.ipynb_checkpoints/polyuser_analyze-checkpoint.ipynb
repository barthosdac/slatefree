```python
import os
import numpy as np
import seaborn as sns
import json
import matplotlib.pyplot as plt
```

```python
def plot_conf_smooth(y,T=None,log=False,title="",color=None,draw_conf=True,padding=False) :
    """
    y of dimension (N,t) :
        N number of experiments
        t number of iteration
    T : size/2 of the smoothing window
    title : title of the plot
    color : color
    draw_conf : if draw the confidence interval
    padding : if the beggining of the smoothed graph has to be padded
        if not the graph start at iteration T
    """
    N,t=y.shape
    mean=y.mean(axis=0)
    
    if T is not None :
        mean=smooth(mean,T=T,padding=padding)
        if padding :
            plt.plot(range(len(mean)),mean,label=title,color=color)
        else :
            plt.plot(range(T,len(mean)+T),mean,label=title,color=color)
    else :
        plt.plot(range(t),mean,label=title,color=color)
    
    if draw_conf :
        conf=y.std(axis=0)/np.sqrt(N)
        if T is not None :
            conf=smooth(conf,T=T,padding=padding)
            if padding :
                plt.fill_between(range(len(conf)), mean-conf, mean+conf, alpha=0.2,color=color)
            else :
                plt.fill_between(range(T,len(conf)+T), mean-conf, mean+conf, alpha=0.2,color=color)
        else :
            plt.fill_between(range(t), mean-conf, mean+conf, alpha=0.2,color=color)
        
    if log :
        plt.xscale('log')
```

```python
def smooth(y,T,padding=False) :
    c=[]
    if padding :
        for i in range(T) :
            c.append(y[i])
    for i in range(T,len(y)-T) :
        c.append(np.sum(y[i-T:i+T])/(2*T))
    return np.array(c)
```

```python
color={'slateFree_msew' : 'g',
       'slateFree_sumw' : 'c',
       'slateQ_w' : '#4444ff',
       'slateFree_mse=' : 'g',
       'slateFree_sum=' : 'c',
       'slateQ_=' : 'k'}
titre={'slateFree_msew' : 'slateFREE(MSE)',
       'slateFree_sumw' : 'slateFREE(SUM)',
       'slateQ_w' : 'slateQ',
       'slateFree_mse=' : 'slateFREE(MSE)',
       'slateFree_sum=' : 'slateFREE(SUM)',
       'slateQ_=' : 'slateQ'}
```

```python
logid='logs30_json'
algos=['slateFree_msew',
       'slateFree_sumw',
       'slateQ_w',
       'slateFree_mse=',
       'slateFree_sum=',
       'slateQ_=']
algos=['slateFree_mse=',
       'slateFree_sum=',
       'slateQ_=']
folders=[logid+'/'+algo for algo in algos]
```

# I - Reward (logs test)

```python
warming=0 # number of episodes for learning the user at each iteration
          # not taken in the graph of the test
```

```python
#we load the logs for each algorithm and do the mean
dname="./"

nb_episode=[]
len_episode=[]

results_test={}
for algo,folder in zip(algos,folders) :

    f=open(f"{folder}/log_test.json")
    log=json.load(f)
    f.close()

    for iteration in log :
        nb_episode.append(len(iteration['episode_reward']))
        len_episode+=list(iteration['episode_length'])
    results_test[algo]=np.array([np.mean(iteration['episode_reward'][warming:]) for iteration in log])
```

```python
graphs=[['slateFree_mse=','slateFree_sum=','slateQ_=']]
for k,graph in enumerate(graphs) :
    for algo in graph :
        result=results_test[algo]
        plot_conf_smooth(result[None,:],T=10,title=titre[algo],color=color[algo])

    plt.ylabel("Reward", fontsize=15)
    plt.xlabel("User", fontsize=15)
    plt.legend()
    try : os.mkdir(dname+logid)
    except : pass
    
    plt.savefig(dname+logid+'/r'+str(k)+'.png')
    plt.show()
```

```python
results_final={}
for algo,folder in zip(algos,folders) :
    
    f=open(f"{folder}/test_final.json")
    log=json.load(f)
    f.close()

    for iteration in log :
        nb_episode.append(len(iteration['episode_reward']))
        len_episode+=list(iteration['episode_length'])
    results_final[algo]=np.array([np.mean(iteration['episode_reward'][warming:]) for iteration in log])
```

```python
for algo in algos :
    results=results_final[algo]
    print(f"{algo}:\t{results.mean():.1f} ± {2*np.std(results)/np.sqrt(results.size):.1f}")
```
