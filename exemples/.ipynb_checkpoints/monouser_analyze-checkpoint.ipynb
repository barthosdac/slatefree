```python
import os
import numpy as np
import seaborn as sns
import json
import matplotlib.pyplot as plt
```

```python
def plot_conf_smooth(y,T=None,log=False,title="",color=None,draw_conf=True,padding=False) :
    """
    y of dimension (N,t) :
        N number of experiments
        t number of iteration
    T : size/2 of the smoothing window
    title : title of the plot
    color : color
    draw_conf : if draw the confidence interval
    padding : if the beggining of the smoothed graph has to be padded
        if not the graph start at iteration T
    """
    N,t=y.shape
    mean=y.mean(axis=0)
    
    if T is not None :
        mean=smooth(mean,T=T,padding=padding)
        if padding :
            plt.plot(range(len(mean)),mean,label=title,color=color)
        else :
            plt.plot(range(T,len(mean)+T),mean,label=title,color=color)
    else :
        plt.plot(range(t),mean,label=title,color=color)
    
    if draw_conf :
        conf=y.std(axis=0)/np.sqrt(N)
        if T is not None :
            conf=smooth(conf,T=T,padding=padding)
            if padding :
                plt.fill_between(range(len(conf)), mean-conf, mean+conf, alpha=0.2,color=color)
            else :
                plt.fill_between(range(T,len(conf)+T), mean-conf, mean+conf, alpha=0.2,color=color)
        else :
            plt.fill_between(range(t), mean-conf, mean+conf, alpha=0.2,color=color)
        
    if log :
        plt.xscale('log')
```

```python
def smooth(y,T,padding=False) :
    c=[]
    if padding :
        for i in range(T) :
            c.append(y[i])
    for i in range(T,len(y)-T) :
        c.append(np.sum(y[i-T:i+T])/(2*T))
    return np.array(c)
```
